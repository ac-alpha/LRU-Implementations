```python
import matplotlib.pyplot as plt
```

```python
file = open("lru_coun.txt","r")
```

```python
lines = file.readlines()
arr = []
```

```python
for i in lines:
    arr.append(int(i.strip()))
```

```python
print(arr)
```

```python
fig = plt.figure()
plt.plot(range(2,11), arr, 'bo-')
fig.suptitle('LRU Counter Implementation', fontsize=20)
plt.xlabel('Number of Frames', fontsize=18)
plt.ylabel('Number of Faults', fontsize=16)
for i_x, i_y in zip(range(2,11), arr):
    plt.text(i_x, i_y, '({}, {})'.format(i_x, i_y))
plt.savefig('lru_count.png')
```

```python
file.close()
```

```python
file = open("time2.txt","r")
lines = file.readlines()
arr = []
for i in lines:
    arr.append(float(i.strip()))
print(arr)
fig = plt.figure()
plt.bar(range(1, 5), arr, color = ['b','r','y','g'], tick_label = ['Counter', 'Clock', 'Stack', 'Aging'])
fig.suptitle('Average Time Comparison', fontsize=20)
plt.xlabel('Implementation Type', fontsize=18)
plt.ylabel('Time taken to execute(in ms)', fontsize=16)
file.close()
plt.savefig('time.png',pad_inches = 0.3)
```
